```python
import pandas as pd
import numpy as np
```

```python
n = 10000
```

```python
customer_id = ["C" + str(i).zfill(7) for i in range(1, n+1)]
vehicle_id = ["V" + str(i).zfill(7) for i in range(1, n+1)]
```

```python
vehicle_value = [100000] * n
```

```python
dates = pd.date_range(start="2024-01-01", end="2024-12-31")
purchase_dates = np.random.choice(dates, n)
```

```python
tenure = np.random.choice([1,2,3,4], n, p=[0.2,0.3,0.4,0.1])
```

```python
premium = tenure * 100
```

```python
start_date = purchase_dates + pd.Timedelta(days=365)
end_date = start_date + pd.to_timedelta(tenure*365, unit="D")
```

```python
df = pd.DataFrame({
    "Customer_ID": customer_id,
    "Vehicle_ID": vehicle_id,
    "Vehicle_Value": vehicle_value,
    "Premium": premium,
    "Policy_Purchase_Date": purchase_dates,
    "Policy_Start_Date": start_date,
    "Policy_End_Date": end_date,
    "Policy_Tenure": tenure
})
```
